```python
df.drop_duplicates('Order ID').resample('ME',on='Order Date')['Order ID'].nunique()
```

```python
monthly_order_counts=df.drop_duplicates('Order ID').resample('ME',on='Order Date')['Order ID'].nunique()
yearly_order_counts=df.drop_duplicates('Order ID').resample('Y',on='Order Date')['Order ID'].nunique()
plt.figure(figsize=(10, 6))
plt.subplot(1,2,1)
plt.plot(monthly_order_counts.index,monthly_order_counts,marker='o', color='skyblue')
plt.title("Number of Unique Orders per Month", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Unique Orders", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(True)


yearly_order_counts=df.drop_duplicates('Order ID').resample('Y',on='Order Date')['Order ID'].nunique()


plt.subplot(1,2,2)
plt.plot(yearly_order_counts.index,yearly_order_counts,marker='o', color='skyblue')
plt.title("Number of Unique Orders per Month", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Unique Orders", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
annual_delivery_days=df.resample('Y',on='Order Date')['Order Delivery (Days)'].mean()

plt.figure(figsize=(10, 6))
plt.subplot(1,2,1)
plt.plot(annual_delivery_days.index,annual_delivery_days,marker='o', color='skyblue')
plt.title("average delivery days per year", fontsize=16)
plt.xlabel("year", fontsize=12)
plt.ylabel("average delivery days", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(True)
```

```python
df.drop_duplicates('Order ID').groupby('Ship Mode').resample('ME',on='Order Date').size().unstack(level=0)
```

```python
ship_mode = df.drop_duplicates('Order ID').groupby('Ship Mode').resample('ME', on='Order Date').size().unstack(level=0)

plt.figure(figsize=(12,7))
for mode in ship_mode.columns:
    plt.plot(ship_mode.index, ship_mode[mode], marker='o', label=mode)

plt.title("Monthly Orders by Ship Mode", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Orders", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Ship Mode')
plt.grid(True)

# حفظ الرسم
plt.savefig("ship_mode_monthly_orders.png", dpi=300, bbox_inches='tight')
plt.show()
```
